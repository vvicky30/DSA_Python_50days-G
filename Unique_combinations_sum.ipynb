### Coding Exercise: Combinations Sum 2
- Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.
- Each number in candidates may only be used once in the combination.

- Note: The solution set must not contain duplicate combinations.

Example :

Input: candidates = [3,5,2,1,3], target = 7
Output: 
[
[1,3,3],
[5,2]
]

```python
def unique_combinationSum(candidates, target):
    #first sorting elements (as here array will be passed with duplicates)
    candidates.sort() # now duplicates will be adjacent to each other   // TC of sorting : O(nlogn)
    res = []
    def backtrack(index,curr,currSum):
        #base-case: if index goes beyond the length of candidates-array or current-sum exceeds the target number 
        if currSum > target and index > (len(candidates)-1) :
            return   # then we simply return without adding that candidate to current_comb and its current-sum
         #if current_sum equals to target-number then we append that curr_comb to result
        if currSum ==target:
            res.append(curr[:])
            return
        # maintaining hash for checking duplicate
        hash={} 
        # step: now choosing choice in choices  [from backtrack-recursion blueprint of psuedo-code]
        for j in range(index,len(candidates)):        
            # step to : check validity of choice here
            if candidates[j] not in hash:
                hash[candidates[j]] = True  
                curr.append(candidates[j])   #step : choose(choice)   [from backtrack-recursion blueprint of psuedo-code]
                backtrack(j+1, curr, currSum+candidates[j])   # recursive-step  with next -index //as we can't repeat the same candidate here
                curr.pop()     #backtrack step
    backtrack(0,[],0)  # here we initialized backtrack fn by starting index 0, curr as empty '[]' and curr_sum as 0
    return res         

# Example usage:
candidates = [3,5,2,1,3]
target = 7
print(unique_combinationSum(candidates, target))

candidates1 = [2, 3, 4, 3, 5]
target = 8
print(unique_combinationSum(candidates1, target))
```
