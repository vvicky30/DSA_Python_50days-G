### Method-2 : simple strategy to solve the monotonic-Array problem  


### method-2 : time and space complexity 


```python
# now Here i am going to implement the method-2 strategy to solve the monotonic array problem 
# this method identifies wehther input-array is monotonic(monotonic-increasing , monotonic-decreasing and 
# monotonic with same values or monotonic with single element or empty input-array will also be considered monotonic) or not , if monotonic in within any of these scenerio 
# it will return True else false if not monotonic.

def identify_monotonic(array):
    n = len(array)  # taking input-array's length 
    # here we check for empty array and pass it True 
    if n == 0 : return True
    
    # now biegning and end side iterators 
    i, j = 0, n-1
    
    # in case when first ele of array is greater than last element of array ; means it should be monotonic descreasing
    if array[i]> array[j]:
        for k in range(n-1):
            if array[k]<array[k+1]: return False   # if next value in input-array is gtreater than last vlaue then it's not monotonic descreasing 
    
    # in case when first ele of array is lesser than last element of array ; means it should be monotonic increasing
    elif array[i]< array[j]:
        for k in range(n-1):
            if array[k]>array[k+1]: return False   # if next value in input-array is lesser than last vlaue then it's not monotonic increasing 
    
    # in case when first ele of array is equal to the last element of array ; means it should be monotonic array consisting of one same elements
    else:     # for conditions like these array[i] == array[j]:
        for k in range(n-1):
            if array[k] != array[k+1]: return False   # if here next value is not equal to the last value , then it;s not monotonic (of identical element)
    
    return True         
        
print(identify_monotonic([5,4,3,2,1]))   # decresing-monotonic 
print(identify_monotonic([5,4,8,2,1]))  # should be false 
print(identify_monotonic([1,2,3,4,5])) # increasing-monotonic 
print(identify_monotonic([1,-1,-1,-1,-2]))# decresing-monotonic 
print(identify_monotonic([1,-1,-1,-1,4])) # should be false
print(identify_monotonic([1]))# input-array with one element 
print(identify_monotonic([]))# Empty input-array  
print(identify_monotonic([3,3,3,3,3])) # monotonic of identical ele
```
