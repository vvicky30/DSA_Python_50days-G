## Combinations_one problem code implementation of optimizing approach:
### this is how we incorporate the optimizing step [controlling 'j' from 'start' to 'need' indexes]:



```python
# approach-backtract code implementation for solving the combination problem :
def combine(n,k):
    res = []
    def helper(start,curr):
        if len(curr)==k:
            res.append(curr[:])
            return
        for j in range(start,n+1):  # step: for choosing choice in choices [from blue-print of backtrack psuedo-code]
                                   # not need to check whether choice is valid or not  
            curr.append(j)         # step: choose the choice 
            helper(j+1,curr)       # recursive -step
            curr.pop()             # backtract-step
    helper(1,[])           # passing values to function ; initialized with start=1 and empty array which saves each combination  
    return res        
#example-usage  
print(combine(5,3))  # going to return 3-3 combinations of number from 1 to 5 
print(combine(4,2))  # should be [1,2,3,4] = [[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]
```

```python
# optimized approach-backtract code implementation for solving the combination problem :
def combine(n, k):
    res = []
    def helper(start,curr):
        if len(curr)==k:
            res.append(curr[:])
            return
        need = k - len(curr)    # for optimizing -step ; this will used to constrain the j untill the specified-indexes ; 
        #as beyond those indexes k-combinations not feasible
        for j in range(start,n-(need-1)+1):  # step: for choosing choice in choices [from blue-print of backtrack psuedo-code]
                                             # not need to check whether choice is valid or not  
            curr.append(j)                   # step: choose the choice 
            helper(j+1,curr)  # recursive -step
            curr.pop()   # backtract-step
    helper(1,[])        # passing values to function ; initialized with start=1 and empty array which saves each combination  
    return res 
#example-usage  
print(combine(5,3))  # going to return 3-3 combinations of number from 1 to 5 
print(combine(4,2))  # should be [1,2,3,4] = [[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]
```
