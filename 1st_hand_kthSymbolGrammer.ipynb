```python
def binary_seq(n):
    # Base case: at the first level, should return "0"
    if n == 1:  #(as we're expecting i-based indexing ; means if n=1 it should put zero on first row and then if n=2 it should put [that zero as its of row-1 alongwith(concatenate)the complement-value of row-1] on row-2 )
        return "0"
    # Recursive case: build the string up to the nth level
    else:
        return binary_seq(n-1) + binary_seq(n-1).translate(str.maketrans('01', '10'))   # here concatenating the string value...this value will be the complement of the string-value on just previous-row   //here string-value is binary-seq that here we're complementing with the help of ex-or approach :  " .translate(str.maketrans('01', '10')) "

def kth_symbol(n, k):
    # Generate the binary sequence
    bin_seq = binary_seq(n)
    # Return the k-th symbol (1-based indexing; that's why used 'k-1' here else it will goes out of range)
    return int(bin_seq[k-1])   # before returning make sure we convert string-value(binary-num) in to integer 


# test-cases checking 
print(binary_seq(3)) # printing binary sequence for  n=3
print(kth_symbol(3,3))  # printing kth element of binary-seq(3)   ; should be 1 
print("==========================")
print(binary_seq(1))
print(kth_symbol(1,1))
print("==========================")
print(binary_seq(4))
print(kth_symbol(4,8))
```
