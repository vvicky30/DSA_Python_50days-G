```python
# final solution
def monotonic_array(array):
    n = len(array)  # for taking length of this input array 
    temp_aar = []*(n-1)  # temparory array of length n-1 [this array will store whether from current-num to further adjacent-num in input-array number increses or decreses ]
    for i in range(n - 1):   # for loop untill second last element of input array
        if array[i] > array[i + 1]:  # if next adjacent-num is less than current-num
            temp_aar.append(-1)  # append -1 to temparory array
        elif array[i] < array[i + 1]:  # if next adjacent-num is greater than current-num
            temp_aar.append(1)  # append 1 to temparory array   
        else:     # in case of next and current number is same 
            temp_aar.append(0)  # then append zero (no increasing or decreasing in next number )
  
  
    if 1 in temp_aar and -1 in temp_aar:   # checking whether 1 and -1 existed in list simulteneously , it means subsequent value increasing and descresing at the same time in the list (so that would'nt be purely increasing or decreasing monotonic_array )
      return False   # this returned here False 
    else:  # this including all case like list of all 1(increasing-monotonic) , or list of all -1(decreasing-monotonic) or list of 1 and zero(increasing-monotonic) or list of -1 and zero (decreasing-monotonic), empty temparory-array in case if input-array has only one element or if input-array itself is empty  
      return True
```

```python
print(monotonic_array([5,4,3,2,1]))   # decresing-monotonic 
print(monotonic_array([5,4,8,2,1]))  # should be false 
print(monotonic_array([1,2,3,4,5])) # increasing-monotonic 
print(monotonic_array([1,-1,-1,-1,-2]))# decresing-monotonic 
print(monotonic_array([1,-1,-1,-1,4])) # should be false
print(monotonic_array([1]))# input-array with one element 
print(monotonic_array([]))# Empty input-array  
```
