

### explanation
- in the number-system cosisting of +ve and -ve no. either if we move towards the left (towards -ve nums) of the zero , towards their extreame or move towards the right (towards =ve nums) of zero, the squared-values will be increased in both cases

- suppose we have this array with -ve integers and in ascending order :[-3,-2, 4,6,7]

- here we start from index-zero (for -3) and from index last (for 7) and in these extreame right and left sqaured values will be highest and compare their sqaured values : 
- here square of -3 will be 9 ; and squre of 7 will be 49 
- so here 49(square of 7) is highest among both and will be saved on the last indices of new array (in case of sorting new squared array in to ascending order)
- and last iterator will be move backward to the another succesive extreame right value which is here 6
- here iterator on left side will be on same value as of -3

- now we have two values on extreame left and right which is -3 and 6 
- so squared of 6 will be 36 greater than square of -3 which 's 9
- so now 36(squared value of 6) will be saved in second last indices of new squared array
- and iteraotor of right will move backward towards successive extreame value whcih's 4 

- now we have 2 values on extreame right and left is 4 and -3 and so on

[Note]: whichever side squared-value will be greater that side of iterator will be moved further towards successive number. 

```python
def sorted_squared(array):
    n= len(array)  # length var of array 
    new_ar = [0]*n  # intializing new array of length same as of old array here
    i,j= 0, n-1 # creating lefthand side(with indices zero) and righthand side itertor(with indices n-1)
    # now we have to tranverse in to new_ar in reverse order (or from right-hand side to left hand side) for saving bigger squared values among the extreame numbers of old array of left and right side:
    for k in reversed(range(n)) :
        if array[i]**2 > array[j]**2 :   # if extreame left-square greater than extreame right square 
            new_ar[k] = array[i]**2   # assigning extreame left squared value to last indices of new array here
            i += 1  # increaing left interator to succesive number 
        else: # if # if extreame right-square greater than extreame left square
            new_ar[k] = array[j]**2  # assigning extreame right squared value to last indices of new array here
            j -= 1 #deceasing right iterator to succesive number 
    return new_ar



print(sorted_squared([-3,-2,4,6,7]))  # passing array for checking 
        
         
             
    
```
