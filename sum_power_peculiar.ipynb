```python
#my way
def power_sum(array, power =1):
    n = len(array)
    total_sum = 0
    for i in range(n):                 # in this approach base-case for recursion is last valid element in array 
        if type(array[i]) == list:
             
            total_sum += (power_sum(array[i], power+1))**(power+1)     # recursive statement 
        
        else:
            total_sum += array[i]
    
    return (total_sum)   
            
# usage :
arr = [1,2,[3,4],[[2]]]
print(power_sum(arr))
             
            
```

```python
#course way
def power_sum(array,power=1):
    sum = 0
    for i in array:
        if type(i) == list:
            sum += power_sum(i, power+1)
        else:
            sum += i
    return sum**power

arr = [1,2,[3,4],[[2]]]
print(power_sum(arr))
# Example usage:
nested_array = [2, 4, [5, [6], 7]]
total_sum = power_sum(nested_array)
print(total_sum)  # Output should be 51990
```
