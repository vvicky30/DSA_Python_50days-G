```python
def kth_Symbol(n,k):
    #base-condition   (not set n less than 1; as we want 1-indexed rows )  
    if n==1:
        return 0
    # first we have to find out length of the sequence (length of grammar):  has formula 2^(n-1)
    length = 2**(n-1)
    # now mid will be the half of length 
    mid = length//2 
    
    if k <=  mid :
        return kth_Symbol(n-1,k)
    
    else :  
        # if kth greater than mid :k> mid then  we should return not or complement of kth_symbol(n-1, k-mid)  
        return int(not(kth_Symbol(n-1, k-mid)))   # as here not function will gives us value in terms of True and false so we should convert it into 1(True) and 0(false)
        #return 1 - (kth_Symbol(n-1, k-mid))   # we can also take complement like this by subracting it from 1   ; as 1-(1)=0  and 1-(0)= 1

# test-cases checking 
print(kth_Symbol(3,3))  # printing kth element of binary-seq(3)   ; should be 1 
print("==========================")
print(kth_Symbol(1,1))
print("==========================")
print(kth_Symbol(4,8))
      
        
```
