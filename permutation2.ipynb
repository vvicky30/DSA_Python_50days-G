# problem-statement : 
- Permutations 2

Given a collection of numbers, nums, that might contain duplicates, return all possible unique permutations in any order.



Example:

nums = [3,3,2]

Output :

[[3,3,2] , [3,2,3], [2,3,3] ]

# permutations-2  Approach  :


# permutations-2  Approach (psuedo-code) :

# permutations-2  code implementation  :


# permutations-2  time and space complexity (will be same as of permutation-1 approach) :




```python
def permuteUnique(nums):
    result = []
    def permutation(i):
        length = len(nums)
        # base-condition : when start-index rrached the end element of array 
        if i == (length-1):
            return result.append(nums[:])   # then append nums array(copy) to the result 
        
        #here first we make hash table (will be maintained for checking duplicate elements of array)
        hash = {}
        for j in range(i,length):   # j from i(startinf-index) to the end index of array 
            if nums[j] not in hash:    # validity of choice statement // here checking whether element is duplicate or not 
                hash[nums[j]] = True    # if not duplicate will maintain its hash as True at index equals to the value(ele at num[j])
                #now swap the values 
                nums[i], nums[j] = nums[j], nums[i]
                #now recursive calling (permutation) for increamented i
                permutation(i+1)
                nums[i], nums[j] = nums[j], nums[i]  # backtrack statement 
                 
             
    
    
    permutation(0)  # here helper fn permutation intilized by starting-index (i) as zero
    return result


# example usage     
nums = [3, 3, 2]
print(permuteUnique(nums))
    
```
