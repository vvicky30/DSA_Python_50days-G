Coding Exercise ( Permutations)
Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order.

Example 1:

nums = [1,4]

Output :[[1,4],[4,1]]

Example 2:

nums = [1,4,5]

Output :[[1,4,5],[1,5,4],[4,1,5],[4,5,1],[5,1,4],[5,4,1]]

```python
#basics : about fixing values(one by one) and concatinating them with array of remaining unfixed-elements
nums = [1,4,5]   # let's say this array
# generally permutations of this three values will be factorial of 3: 3*2*1= 6
#there will be six combinations  , among which two-two will be of fixing each value and swapping remaining two values in array (like):
# [1,4,5] [1,5,4]  # fixing 1 
# [4,1,5] [4,5,1]  # fixing 4
# [5,1,4] [5,4,1]  # fixing 5
 


# this gives idea about how slicing of remaining values fo arrays and concatenating them with fixed-ele:-  
# formula : [nums[i]] + nums[:i] +nums[i+1:]    // with for loop

print([nums[0]]+nums[:0] +nums[1:]) # here we fixed element at zero-th index(first-ele)   ; if i=0   which's here '1'  
print([nums[1]]+nums[:1] +nums[2:]) # here we fixed element at one index(second-ele)   ; if i=1       which's here '4'
print([nums[2]]+nums[:2] +nums[3:]) # here we fixed element at two index(third-ele)   ; if i=2        which's here '5'     
#[Note: while concatenating fixed-ele with remaining-elements(in form of array already due to slicing ) we have to convert fixed-ele to list first]
```

```python
def permute(nums):
    #what would be the base-conition here //as we know that i/p-array can't be empty , it contains atlest 1 element
    #base-conition of recursion here  // if array contains signle element let's say :[3] then we have to return it in array of arrays like this [[3]]
    length = len(nums)
    op = []  #intializing o/p list here which contains all possible permuted arrays   
    if length == 1:
        return [nums]   # in array of arrays
    
    # if nums arrays has more than 1 element like this :[4,5,6]
    # and we want to permute all the possible cases : like firsrtly here fixing '4' and permutate "5,6" like this : "5,6" & "6,5" here we modify the state of problem at hand in-places (in-places : 1th and 2nd index)[while making ele at zeroth index fix], and concatenate all these two permutation "5,6" & "6,5" to fixed ele(here which's '4') like this : '4,5,6' and 4,6,5                 
    # and will do the same by fixing 5 and permutate '4,6' and further fixing 6 and permutate '4,5'
    for i in range(length):
        fixed_ele = nums[i]   # fixing current ele
        unfixed_ele = nums[:i] + nums[i+1:]    # remaining elements after current ele
        #now we will permuate unfixed_ele(all those remaining elements will be passed in arrays as arugments to this recursive-fn)
        
        permutations = permute(unfixed_ele)
        
        for permt in permutations:
             # here we cancatenate all possible permutations of unfixed_ele to the fixed ele (current-ele)  //note: convert current-ele to list :
            op.append([fixed_ele] + permt)
    
    return op


# example usage     
nums = [1, 4, 5]
print(permute(nums))
```

```python
# other test-cases:
n1= [5,6,8,2]   # 4 elements so total permutations should be 4 factorial : 24
print(permute(n1))
p = permute(n1)
print(len(p))
```
