```python
def findthewinner(n,k):
    def last_standing(n):
        
        #base condition
        if n==1 : return 0  
          # if n=1 means last element is remaining then we would return zero simply (as following zero-based indexing)
        else: # for n greater than 1 
            # here we will call this same function recursively upto the n ==1 
            return ( last_standing(n-1) + k ) % n
        
    return last_standing(n)+1    # returning fn(n)+1 index as if returning in 1-based index
```

```python
#cheking for test-cases here 
print(findthewinner(1,3))  # should be 1
print(findthewinner(1,4))
print(findthewinner(2,3))  # should be 2
print(findthewinner(3,3))  # should be 2
print(findthewinner(4,3))  # should be 1
print(findthewinner(41,2))  # should be 19   //based on historic scenerio of josephus and its 40 friends trapped in cave surrounded by roman-soldiers
                                                    
```
