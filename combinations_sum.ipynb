# Combinations_sum problem statement:

### Combinations Sum 1
- Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.
- The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the frequency of at least one of the chosen numbers is different.
- (the integers in the candidates array are all non negative )

Example 1:

Input: candidates = [2,3,8,9], target = 9

Output: [[2,2,2,3],[3,3,3],[9]]

Explanation:

2 and 3 are candidates, and 2 + 2 + 2+ 3 = 9. Note that 2 can be used multiple times.

3 is a candidate and 3+3+3 = 9.

9 is a candidate, and 9 = 9.

These are the only 3 combinations.

# Combinations_sum problem approach [code implementation] :


```python
def combinationSum(candidates, target):
    res = []
    def backtrack(index,curr,currSum):
        #base-case: if index goes beyond the length of candidates-array or current-sum exceeds the target number 
        if currSum > target or index > (len(candidates)-1) :
            return   # then we simply return without adding that candidate to current_comb and its current-sum
         #if current_sum equals to target number then we append that curr_comb to result
        if currSum ==target:
            res.append(curr[:])
            return
        # step: now choosing choice in choices  [from backtrack-recursion blueprint of psuedo-code]
        for j in range(index,len(candidates)):        
            # no need of step to check validity of choice here
            curr.append(candidates[j])   #step : choose(choice)   [from backtrack-recursion blueprint of psuedo-code]
            backtrack(j,curr,currSum+candidates[j])   # recursive-step
            curr.pop()     #backtrack step
    backtrack(0,[],0)  # here we initialized backtrack fn by starting index 0, curr as empty '[]' and curr_sum as 0
    return res         

# Example usage:
candidates = [2, 3, 6, 7]
target = 7
print(combinationSum(candidates, target))

candidates1 = [2, 3, 8, 9]
target = 9
print(combinationSum(candidates1, target))
```

```python
'''
Step-by-Step Trace
Initial Call to backtrack:

0.  backtrack(0, [], 0) is called.
index = 0, current_combination = [], current_sum = 0.


1. First Level of Recursion:
Include candidates[0] which is 2.
current_combination = [2], current_sum = 2.
Call backtrack(0, [2], 2).

2. Second Level of Recursion:
Include candidates[0] which is 2 again.
current_combination = [2, 2], current_sum = 4.
Call backtrack(0, [2, 2], 4).

3.Third Level of Recursion:
Include candidates[0] which is 2 again.
current_combination = [2, 2, 2], current_sum = 6.
Call backtrack(0, [2, 2, 2], 6).

4.Fourth Level of Recursion:
Include candidates[0] which is 2 again.
current_combination = [2, 2, 2, 2], current_sum = 8.
Since current_sum > target, return without adding to the result.
Backtrack by removing the last added 2, making current_combination = [2, 2, 2] ; and current sum =6.

5. Try Next Candidate at Third Level:
Exclude candidates[0].
Try candidates[1] which is 3.
current_combination = [2, 2, 3], current_sum = 7.
Call backtrack(1, [2, 2, 3], 7).

6. Valid Combination Found:
current_sum == target, so [2, 2, 3] is added to result.

7.  Backtrack Further:
Remove 3, making current_combination = [2, 2]; current sum = 4
Try candidates[2] which is 6.
current_combination = [2, 2, 6], current_sum = 10.
Since current_sum > target, return.

8. Continue Backtracking:
Remove 6, making current_combination = [2, 2].
Try candidates[3] which is 7.
current_combination = [2, 2, 7], current_sum = 11.
Since current_sum > target, return.

9. Backtrack to Second Level:
Remove the last 2, making current_combination = [2].

10.Continue with other candidates at each level similarly.
-------------
-------------
Skipping Some Steps, Final Combination:
Ultimately, trying 7 alone will result in current_combination = [7].
current_sum == target, so [7] is added to result.
Final Result
'''
#After all the recursive calls and backtracking steps, the function will have found all valid combinations that sum up to 7. The final result will be:
# [[2, 2, 3], [7]]
```
