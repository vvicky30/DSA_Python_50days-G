```python
def findTheWinner(n, k):
    # base-case
    if (n==1): return 1   # if in the cicrlce only one element is there which;s '1' here then whatevder will be the 'k' it will it will return 1 as index here [note: we're using 1-based index here in the solution ]
    else: # if n is greater than 1 
    # then for each counting(clock-wise) untill the k one elel=ment will be deleted 
    #hence with each counting(clock-wise) untill the k n will be less by one ; (n-1) whereas k will be constant 
    # for finding out which element deleted in each counting we have this formula:
    # (fn(n-1,k) + k-1) % size ; here as size we use (n+1) instead of n-1; here we used 'k-1' 'coz we are working with zero-based indexing
    # now above formula is recusrsive-statement as we can see    
         return (findTheWinner(n-1,k) + k-1) % n+1


#[note:] whenever we encounter problem taking place in circular fassion we should find the solution using modulo(%) formula , in such scenerios 
```
