# Approach-3 Code-implementation


```python
def findthewinner(n,k):
    # fisrstly initialized last survivior as zero  as if n=1 ; means only one element remaining it will return simply 1 as index(in terms of 0-based index)
    last_survivior = 0
    # now we make loop from (n>1)=2 to n+1    
    for i in range(2, n+1):
        last_survivior = (last_survivior + k) % i     # for here each i acts as increamenting 'n' 
        
    return last_survivior +1 #(as for returning in terms of 1-based indexing)
```

```python
#cheking for test-cases here 
print(findthewinner(1,3))  # should be 1
print(findthewinner(1,4))
print(findthewinner(2,3))  # should be 2
print(findthewinner(3,3))  # should be 2
print(findthewinner(4,3))  # should be 1
print(findthewinner(41,2))  # should be 19   //based on historic scenerio of josephus and its 40 friends trapped in cave surrounded by roman-soldiers
                                                    
```
