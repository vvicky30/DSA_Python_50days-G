# superset problem statement :

- Subsets
- Question:
Power Set - Given an integer array of unique elements, return all possible subsets (the power set). The solution set must not contain duplicate subsets. Return the solution in any order.

# superset problem test-cases :



```python
def power_set (nums):
    # if array has no element simply the array of one array(blank) will be returned 
    # //this one blank array will be returned in any conditions(even if array has elements) as a power_set along with other power_sets
    # so we should start result with minimum one-blank array 
    result = [[]]
    length = len(nums)  # length of the array 
    if length == 0:
        # if array is empty
        return result   # simply return result which is intialized with blank array
    for i in range(length):
        result += [subset + [nums[i]] for subset in result]   # then we append combinations of current element of array with already appended(existed) subsets in result.   //using list comprehension technique here 
                                                                 
    return result

# Example usage:
empty_nums = []
print(power_set(empty_nums))
nums = [1,2,3,4]
print(power_set(nums))


'''
if array has zero length means empty array passed as above we passed empty_array
 - then function will return reault(initialized with array of one-blank array) as it's 

but if array passed is not empty for ex: nums = [1,2,3,4]
-then function first iterate over passed array one by one and append combinations of current element with already appended(existed) subsets in result to the result.

if i = 0 means ele = 1 
result  += [[]+[1]] = [1]   
result will be [[],[1]]

if i = 1 means ele = 2 
result  += [[]+[2]] = [2]   //step in list comprehension
result  += [[1]+[2]] = [1,2]   //step in list comprehension
result will be [[],[1],[2],[1,2]]

if i = 2 means ele = 3
result  += [[]+[3]] = [3]   //step in list comprehension
result  += [[1]+[3]] = [1,3]   //step in list comprehension
result  += [[2]+[3]] = [2,3]   //step in list comprehension
result  += [[1,2]+[3]] = [1,2,3]   //step in list comprehension  
result will be [[],[1], [2], [1,2], [3], [1,3], [2,3], [1,2,3]

if i = 3 means ele = 4
result  += [[]+[4]] = [4]   //step in list comprehension
result  += [[1]+[4]] = [1,4]   //step in list comprehension
result  += [[2]+[4]] = [2,4]   //step in list comprehension
result  += [[1,2]+[4]] = [1,2,4]   //step in list comprehension
result  += [[3]+[4]] = [3,4]   //step in list comprehension
result  += [[1,3]+[4]] = [1,3,4]   //step in list comprehension
result  += [[2,3]+[4]] = [2,3,4]   //step in list comprehension
result  += [[1,2,3]+[4]] = [1,2,3,4]   //step in list comprehension
result will be [[], [1], [2], [1, 2], [3], [1, 3], [2, 3], [1, 2, 3], [4], [1, 4], [2, 4], [1, 2, 4], [3, 4], [1, 3, 4], [2, 3, 4], [1, 2, 3, 4]]
'''
```
