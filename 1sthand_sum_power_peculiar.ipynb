```python
# first we tries to implelemt how we can add the elements of array usign recursion 
def rec_arr_sum(array):
    n = len(array)
    # base case / if length of array is zero ; means if array has no element then its sum will be zero 
    if n== 0: return 0
    
    # if array has some elements then sum of all those will be zero-th ele of array + sum of remaining elements in array 
       # for sum of reamianing elements in array ; we will call this fn recusively for each and every recursive elements to till no other remaining element but sum of all
                                                                            
    return array[0] + rec_arr_sum(array[1:])  

#let say this is array with this much elements:
array = [3,4,5,6]
print(rec_arr_sum(array))
```

```python
# code impelentation for solving sum of all elements in the peculiar array , raising each power of sum to their respective levels of nesting 
def power_sum(array,power=1):  # as here we can see power's default value is 1
# intitialized total_sum here
  total_sum = 0
  length = len(array)
  for i in range(length):
    if isinstance(array[i], list):  # if interator found another array in the array instead of element
      # then it recursively call the function on it.
                       # if found nested array increase powers by one in functional argument below
      total_sum += (power_sum(array[i],power+1))**(power+1)
    else:
      # Otherwise, add the element to the total sum.
      total_sum += array[i]

  return total_sum


# Example usage:
nested_array = [2, 4, [5, [6], 7]]
total_sum = power_sum(nested_array)
print(total_sum)  # Output should be 51990

#as here in innesrmost nested array it has ele = '6'  so    2 + 4 ( 5 + (0+6)^3 + 7 )^2
'''
step by step [2 + 4 ( 5 + (0+6)^3 + 7 )^2] (hit & trail):
first is the int ele 2 : 0+2 ; total_sum=2
second is the int ele 4 : 2+4 ; totl_aum = 6
third is the list-ele (power raised = 1+1=2) : 6+ power_sum([5,[6].7], 2):
   6 +               first is the int-ele 5 : 0 +5 ; total_sum =5
   6 +               second is the list-ele  (power-raised = 2+1 =3) : 5+  power_sum([6], 3):
   6 +                                          (   5+                  first is the int-ele 6 :  (6)^3    )^2
   6 +               third is the int-ele 7     (   221 + 7 )^2
   6 +               51984
   51990                                                                                            [finally]
                
'''
```

```python
def power_sum(array, power=1):
    total_sum = 0
    length = len(array)
    
    for i in range(length):
        if isinstance(array[i], list):
            # If an element is a list, recursively call power_sum on it with incremented power
            total_sum += (power_sum(array[i], power + 1))**(power + 1)
        else:
            # Otherwise, add the element to the total sum
            total_sum += array[i]

    return total_sum

# Usage
arr = [1, 2, [3, 4], [[2]]]
print(power_sum(arr))

# Example usage:
nested_array = [2, 4, [5, [6], 7]]
total_sum = power_sum(nested_array)
print(total_sum)  # Output should be 51990
```
