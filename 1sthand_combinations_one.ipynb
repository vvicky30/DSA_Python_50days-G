- Coding Exercise: Combinations
- Given two integers n and k, return all possible combinations of k numbers chosen from the range [1, n].

- You may return the answer in any order.

Example:

n = 4

k=2

Output =

[

[1,2],[1,3],[1,4],

[2,3],[2,4],

[3,4]

]

```python
def combinations(n, k):
    result = [] # result where we going to save all combinations 
    def backtrack_comb(start, combination):
        if len(combination) == k:  #meaning if combination's length equals to 'k' then we have to append that combination to the result 
        #//ex: if [1,2,3] then append this combination to result as its length equals to 3 
            result.append(combination[:])
            return
        for i in range(start, n + 1):   # for loop for iterating over array 
            combination.append(i)       # appending each element from 1 to 5 on 1st level of recursion ; 
            #then recrsively appending values // means appending values from 2 to 5 on 2nd level of recursion and so on...
            backtrack_comb(i + 1, combination)  #recursion -step  
            combination.pop() # backtrack-step
    backtrack_comb(1, [])   # passing values to function ; initialized with start=1 and empty array which saves each combination  
    return result
#exampl-usage  
print(combinations(5,3))  # going to return 3-3 combinations of number from 1 to 5 
print(combinations(4,2))  # should be [1,2,3,4] = [[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]
```

```python
'''

///  Example Execution (n = 5, k = 3):

Initial Call: backtrack(1, [])

start = 1, combination = []
First Level:

Loop from i = 1 to 5:
i = 1:
combination = [1]
Call backtrack(2, [1])
i = 2:
combination = [2]
Call backtrack(3, [2])
i = 3:
combination = [3]
Call backtrack(4, [3])
i = 4:
combination = [4]
Call backtrack(5, [4])
i = 5:
combination = [5]
Call backtrack(6, [5]) (ends since start > n)

Second Level:

For backtrack(2, [1]):
Loop from i = 2 to 5:
i = 2:
combination = [1, 2]
Call backtrack(3, [1, 2])
i = 3:
combination = [1, 3]
Call backtrack(4, [1, 3])
i = 4:
combination = [1, 4]
Call backtrack(5, [1, 4])
i = 5:
combination = [1, 5]
Call backtrack(6, [1, 5])

Third Level:

For backtrack(3, [1, 2]):
Loop from i = 3 to 5:
i = 3:
combination = [1, 2, 3]
Valid combination found, add [1, 2, 3] to result
i = 4:
combination = [1, 2, 4]
Valid combination found, add [1, 2, 4] to result
i = 5:
combination = [1, 2, 5]
Valid combination found, add [1, 2, 5] to result
This process continues, exploring all possible combinations of 3 elements out of 5.
'''
#o/p: [[1, 2, 3], [1, 2, 4], [1, 2, 5], [1, 3, 4], [1, 3, 5], [1, 4, 5], [2, 3, 4], [2, 3, 5], [2, 4, 5], [3, 4, 5]]
```
