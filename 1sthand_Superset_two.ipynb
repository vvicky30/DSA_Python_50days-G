Given an integer array nums that may contain duplicates, return all possible

subsets

(the power set).

The solution set must not contain duplicate subsets. Return the solution in any order.

Example : 

nums = [1,3,3]

Output =

[

[],

[1],

[3],

[1,3],

[3,3],

[1,3,3]

]

```python
def subsetsWithDup(nums):
    nums.sort()  # Sort the array to handle duplicates //as duplicates will be adjacent to each other
    # if array has no element simply the array of one array(blank) will be returned //this one blank array will be returned in any conditions(even if array has elements) as a power_set along with other power_sets
    # so we should start result with minimum one-blank array
    result = [[]]
    length = len(nums) # length

    if length == 0:
        # if length of array is zero ; means empty array 
        return result # will return 

    for i in range(length):
      if i==0 or nums[i] != nums[i-1]:  # cheking duplicate ; if current element is not same as last element (provided that current element shoud'nt be first element)
         length_before = len(result)  # going to take length of result where subsets already existed
      result += [subset + [nums[i]]  for subset in result[len(result) - length_before:]]
                         # then we append combinations of current element of array with already appended(existed) subsets in result. 
                          #[Note here chatch is  we only make combinations of current elements with last newly added subsets in the result (provided that if element is duplicate)]  
                          # that's why we used this slicing of result in the list comprehension technique : result[len(result) - length_before:]
                          #//using list comprehension technique here
    return result  

nums1 = [1,3,3]
print(subsetsWithDup(nums1)) 
nums2 = [1,3,3,4,4]            
print(subsetsWithDup(nums2)) 

'''
Hit and trail for above example : [1,3,3]

i = 0 means ele = 1    //here i ==0 means first element of array and it's not duplicate
length_before  = 1

result[len(result) - length_before:]] =  1-1 = 0   ; 0:  from zero it will traverse in result
result  += [[]+[1]]    = [1]

result will be [[],[1]]

i = 1 means ele = 3  // here 3 is not equals to last element of array '1' 
length_before  = 2

result[len(result) - length_before:]] =  2-2 = 0   ; 0:  from zero it will traverse in result
result  += [[]+[3]] = [3] 
result  += [[1]+[3]] = [1,3]

result will be [[],[1], [3], [1,3]]

i = 2 means ele = 3  // here 3 is duplicate ; and its equal to last element of array '3' 
length_before  = 2    //length will be same as previous length   

result[len(result) - length_before:]] =  4-2 = 2   ; 2:  from two it will traverse in result
result  += [[3]+[3]] = [3,3]
result  += [[1,3]+[3]] = [1,3,3]

result will be [[],[1],[3], [1,3], [3,3], [1,3,3]]
'''
```
